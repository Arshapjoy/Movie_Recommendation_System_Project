### importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### loading the data

```python
reviews=pd.read_csv("/content/rotten_tomatoes_movie_reviews.csv")
movies=pd.read_csv("/content/rotten_tomatoes_movies.csv")
```

```python
reviews.info()
```

```python
movies.info()
```

```python
combined_df = pd.merge(reviews, movies, on='id', how='outer')
```

```python
missing_values = combined_df.isnull().sum()
```
